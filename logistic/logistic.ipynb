逻辑回归：根据现有数据对分类边界线建立回归公式，以此进行分类。

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
def load_data(path):
    datasets = pd.read_csv(path,sep='\t',header=None)
    data = datasets.iloc[: , :-1]
    label = datasets.iloc[: , -1]
    return data,label
```

```python
path = '/Users/yangyang/Desktop/GitHub/Deep-Learning/logistic/testSet.txt'
data,label = load_data(path)

# 原来label是(100,) 增加一维后(100,1)
label = label[ : ,np.newaxis]
```

```python
label0 = np.where(label.ravel() == 0)
plt.scatter(data.values[label0,0],data.values[label0,1],marker='x',color='r',label = '0')
label1 = np.where(label.ravel() == 1)
plt.scatter(data.values[label1,0],data.values[label1,1],marker='x',color='b',label = '1')
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc = 'upper left')
plt.show()
```

```python
def sigmod(x):
    return 1.0 / (1 +np.exp(-x))

#批梯度下降ngd
def bgd(x,y,learn_rate = 0.001,iters = 1000):
    m,n = x.shape
    weight = np.zeros((n,1))   #初始化为0
    weights = np.zeros((iters,n))  #记录权重的变化
    for i in range(iters):
        out = sigmod(x.dot(weight))
        error = y - out  #(100*3)
        dw = -x.T.dot(error)
        weight += -learn_rate * dw
        weights[i,:] = weight.ravel().T
    return weight,weights
```

```python
m,n = data.shape
X = np.hstack((np.ones((m,1)),data.values))
weight,weights = bgd(X,label)
```

对于线性回归的问题来说，目的是找到一条能够线性拟合数据的函数来更好的拟合数据。
使用sigmoid函数是为了说明得出的分类结果属于此类结果的结果概率有多大
最优解：得出的属于1分类的概率应该尽可能的大；同理属于0分类的概率也尽可能大
sigmoid表示的是结果概率

```python
label1 = np.where(label.ravel() == 0)
plt.scatter(data.values[label1,0],data.values[label1,1],marker='x',color = 'r',label = '0')
label2 = np.where(label.ravel() == 1)
plt.scatter(data.values[label2,0],data.values[label2,1],marker='o',color = 'b',label = '1')
xx = np.arange(-4.0,3.0,0.1)
yy = (-weight[0] - weight[1]*xx) / weight[2]
plt.plot(xx,yy,"y_")
plt.xlabel('x1')
plt.ylabel('x2')
plt.legend(loc = 'upper left')
plt.show()
```

随机梯度下降sgd

```python
def sgd(x,y,learn_rate=0.01,iters = 200):
    m,n = x.shape
    weight = np.zeros((n,1))  #new一个(3*1)矩阵是为了做矩阵乘法
    weight0 = []
    weight1 = []
    weight2 = []
    for i in range(iters):
        for j in range(m):
            out = sigmod(x[j,:].reshape(1,-1).dot(weight))  #(1*1)
            error = y[j,:].reshape(1,-1) - out
            dw = -x[j,:].reshape(1,-1).T.dot(error)
            weight += -learn_rate * dw
            weight0.append(weight[0,0])
            weight1.append(weight[1,0])
            weight2.append(weight[2,0])
    return weight,weight0,weight1,weight2
            
```

```python
weight,weight0,weight1,weight2 = sgd(X,label)
print(weight)
```

```python
#来看看参数weight的随着迭代次数的收敛情况,可以看到有上下波动的情况，但是整体还是逐渐收敛的
plt.subplot(311)
plt.plot(weight0,"b")
plt.ylabel("w0")

plt.subplot(312)
plt.plot(weight1,"r")
plt.ylabel("w1")

plt.subplot(313)
plt.plot(weight2,"y")
plt.ylabel("w2")
plt.xlabel("iters")
plt.show()
```
